# Gencode, SJ.out, and eXpress Processing

I'm going to make some useful files using the Gencode annotation 
for downstream use. I'm also going to combine the SJ.out files 
from STAR and the results files from eXpress. These files can take
several minutes to make but we only need to do it once. 

```python
import glob
import os

import cdpybio as cpb
import ds2014 as ds
import pandas as pd
```

```python
curdir = os.getcwd()
outdir = os.path.join(os.path.split(curdir)[0], 
                      'output', 
                      'gencode_sjout_express_processing')
ds.makedir(outdir)

meta = pd.read_table('../data/metadata.tsv',
                     index_col=0, header=0)
```

## Gencode

We'll parse the Gencode GTF file and make three files so that we can have easier to access to some information later on.

#### `gene_info.tsv`

A table with the Gencode gene ID, gene type, gene status, and gene name.

#### `transcripts_genes.tsv`

A table with Gencode transcript IDs and the corresponding Gencode gene ID for the gene that contains that transcript.
#### `junction_info.tsv`

A table with information about the splice junctions implied by the GTF file. Has the following columns: gene (Gencode gene ID), chrom, start, end, strand, chrom:start, chrom:end, donor, acceptor, intron. Note that the coordinates in this file are one-based and that splice junctions are represented by the one-based coordinates of the first and last base pairs of the junction's intron.

It can take several minutes to make these files, but we only have to do it once.

```python
# Get gene info
f = os.path.join(outdir, 'gene_info.tsv')
if not os.path.exists(f):
    print('Making '.format(f))
    gene_info = cpb.gencode.make_gene_info_df(ds.gencode_gtf)
    gene_info.to_csv(f, sep='\t')
else:
    gene_info = pd.read_table(f, index_col=0,
                              header=0)
print(f)
print(gene_info.head())

# Get mapping from transcripts to genes
f = ds.transcripts_genes
if not os.path.exists(f):
    print('Making '.format(f))
    tg = cpb.gencode.make_transcript_gene_se(ds.gencode_gtf)
    tg.to_csv(f, sep='\t')
else:
    tg = pd.read_table(f, index_col=0,
                       header=0)
print(f)
print(tg.head())
    
# Get splice junction information
f = ds.junction_info
if not os.path.exists(f):
    print('Making '.format(f))
    jxn_info = cpb.gencode.make_splice_junction_df(ds.gencode_gtf)
    jxn_info.to_csv(f, sep='\t')
else:
    jxn_info = pd.read_table(f, index_col=0,
                             header=0)
print(f)
print(jxn_info.head())
```

## eXpress

Now we'll combine the eXpress output for each sample. We'll make a file that represents gene expression using the mapping of transcripts to genes that we made from the Gencode GTF. The expression of a gene will be defined as the sum of the expression values of the gene's different transcripts.

```python
def define_sample_name(x):
    """Convert filename into sample name"""
    return os.path.split(x)[1].split('_')[0]

f = ds.transcript_eff_counts
g = ds.gene_eff_counts
if (not os.path.exists(f)) or (not os.path.exists(g)):
    print('Making {} and {}'.format(f, g))
    fns = glob.glob(os.path.join(ds.expressdir,'*xprs'))
    t_counts, g_counts = cpb.express.combine_express_output(fns,
                                                            tg=ds.transcripts_genes,
                                                            define_sample_name=define_sample_name)
    t_counts.to_csv(f, sep='\t')
    g_counts.to_csv(g, sep='\t')
else:
    t_counts = pd.read_table(f, index_col=0, header=0)
    g_counts = pd.read_table(g, index_col=0, header=0)
    
print(f)
print(t_counts.head().T.head().T)
print(g)
print(g_counts.head().T.head().T)

f = ds.transcript_fpkm
g = ds.gene_fpkm
if (not os.path.exists(f)) or (not os.path.exists(g)):
    print('Making {} and {}'.format(f, g))
    fns = glob.glob(os.path.join(ds.expressdir,'*xprs'))
    t_counts, g_counts = cpb.express.combine_express_output(fns, column='fpkm',
                                                            tg=ds.transcripts_genes,
                                                            define_sample_name=define_sample_name)
    t_counts.to_csv(f, sep='\t')
    g_counts.to_csv(g, sep='\t')
else:
    t_counts = pd.read_table(f, index_col=0, header=0)
    g_counts = pd.read_table(g, index_col=0, header=0)
    
print(f)
print(t_counts.head().T.head().T)
print(g)
print(g_counts.head().T.head().T)
```

## STAR

Now we'll combine the `SJ.out.tab` files from STAR. This relies on the [cdpybio python package](https://github.com/cdeboever3/cdpybio). First we combine all of the different `SJ.out.tab` files so that we have a file with the coverage of every junction by uniquely mapped reads. Then we filter novel junctions to keep only those that share a donor or acceptor (or both) with a known Gencode junction.

### `Log.final.out` files

We'll combine together the `Log.final.out` files so we
can look at some alignment statistics.

```python
def define_sample_name(x):
    """Convert filename into sample name"""
    return os.path.split(x)[1].split('.')[0]

if not os.path.exists(ds.star_logs):
    logs = cpb.star.make_logs_df(glob.glob('{}/*out'.format(ds.logfinaldir)),
                                 define_sample_name=define_sample_name)
    logs.to_csv(ds.star_logs, sep='\t')
else:
    logs = pd.read_table(ds.star_logs, header=0, index_col=0)
print(ds.star_logs)
print(logs.head())
```

### `SJ.out.tab` files

#### Junction filtering

##### Unique read coverage

`SJ.out.tab` files contain many unannotated junctions with low coverage. Many of these are likely noise or cannot be used in downstream analyses due to the low coverage. When combining together the `SJ.out.tab` files, we will only keep junctions that have at least 20 uniquely mapped junctions-spanning reads summed over all samples. So if a junction has 20 unique junction-spanning reads in one sample, we keep it. If a junction has one unique junction-spanning read in 20 different samples, we keep it. Other junctions are discarded. Since some analyses will only focus on a particular subset of the cancers I'm including in the study, we'll combine `SJ.out.tab` for each different subset since the coverage filtering will behave differently for different groups of samples.

##### Novel junctions

There are many novel (potential) junctions in the `SJ.out.tab` files. To enrich for novel junctions that are likely real (as opposed to technical noise or something) we'll only keep novel junctions that share a donor acceptor with a known Gencode junction. Note that both a novel junction's donor and acceptor could be in Gencode but not annotated together as a splice site.

```python
def define_sample_name(x):
    """Convert filename into sample name"""
    return os.path.split(x)[1].split('.SJ')[0]

def combine_sj_out_files(cancers, meta=meta, outdir=outdir, 
                         define_sample_name=define_sample_name):
    """Run cpb.star.combine_sj_out for all samples for a subset of cancers"""
    samples = list(meta[meta.disease_abbr.apply(lambda x: x in cancers)].index)
    fns = [ os.path.join(ds.sjoutdir, '{}.SJ.out.tab'.format(x)) for x in samples ]
    sf = os.path.join(outdir, '{}_combine_sj_out_stats.txt'.format('_'.join([ x.lower() for x in cancers ])))
    counts, annot, stats = cpb.star.combine_sj_out(fns, 
                                                   ds.junction_info,
                                                   define_sample_name=define_sample_name)
    return counts, annot, stats
```

```python
cancers = ['UM']
f = ds.um_sj_counts
g = ds.um_sj_annot
h = ds.um_sj_stats
if (not os.path.exists(f)) or (not os.path.exists(g)) or (not os.path.exists(h)):
    print('Combining junctions for {}'.format(', '.join(cancers)))
    counts, annot, stats = combine_sj_out_files(cancers)
    counts.to_csv(f, sep='\t')
    annot.to_csv(g, sep='\t')
    f = open(h, 'w')
    f.write('\n'.join(stats) + '\n')
    f.close()
else:
    counts = pd.read_table(f, index_col=0, header=0)
    annot = pd.read_table(g, index_col=0, header=0)
    t = open(h)
    stats = [ x.strip() for x in t.readlines() ]
    t.close()
print(f)
print(counts.head())
print(g)
print(annot.head())
print(h)
print('\n'.join(stats))
```
