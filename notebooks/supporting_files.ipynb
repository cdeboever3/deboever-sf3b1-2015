# Supporting Files

This notebook creates the supplementary files from the paper.

```python
import os

import ds2014 as ds
import pandas as pd
```

```python
curdir = os.getcwd()
outdir = os.path.join(os.path.split(curdir)[0], 
                      'output', 
                      'supporting_files')
ds.makedir(outdir)
```

```python
meta = pd.read_table(ds.metadata, header=0,
                     index_col=0)

gene_info = pd.read_table(ds.gene_info, header=0,
                          index_col=0)

# cancer census genes from COSMIC
cancer_gene_census = pd.read_table(ds.cancer_gene_census,
                                   index_col=0)
# cancer genes from NCG4.0 (http://ncg.kcl.ac.uk/)
cancer_genes = pd.read_table(ds.cancer_genes,
                             index_col=1, header=0)
# tumor suppressors from TSGene (http://bioinfo.mc.vanderbilt.edu/TSGene/)
ts_genes = pd.read_table(ds.ts_genes,
                         index_col=1, header=0)
```

## File S1

Samples used in this study. SF3B1 mutated samples have 
columns for frequency of SF3B1 mutation in RNA-seq data, 
mutation type, codon change and whether the mutation is 
in the HEAT 5-9 repeats. These columns are empty for 
SF3B1 wild-type tumor samples.

```python
sf3b1_mut_info = pd.read_table(ds.sf3b1_mut_info,
                               header=0, index_col=None)
file_s01 = sf3b1_mut_info[['name', 'disease_abbr',
                           'alt_freq', 'mut_type',
                           'codon_change', 'HEAT']]
file_s01['alt_freq'] = file_s01.alt_freq.apply(lambda x: '{:.2f}'.format(x)).values
file_s01.columns = ['Sample Name', 'Cancer', 'Alternate Allele Frequency',
                    'Mutation Type', 'Codon Change', 'In HEAT 5-9']
file_s01['SF3B1 mutated'] = True
wt = pd.DataFrame(columns=file_s01.columns)
wt['Sample Name'] = meta[meta.condition == 'WT'].index
wt['SF3B1 mutated'] = False
wt['Cancer'] = meta.ix[wt['Sample Name'], 'disease_abbr'].values
file_s01 = pd.concat([file_s01, wt])
file_s01.to_csv(os.path.join(outdir, 'file_s01.tsv'),
                sep='\t', index=False)
print('{} rows'.format(file_s01.shape[0]))
file_s01.head()
```

## File S2

Summary of differential junction usage results from DEXSeq. 
DEXSeq was used to test for differential splice junction 
usage in a joint analysis of the CLL, BRCA, and UM samples 
as well as individually for each cancer type. “Novel” indicates 
that the junction is not annotated in Gencode. Proximal 
indicates that a novel 3’SS is 10-30 bp upstream of a canonical 
Gencode 3’SS. In this paper, we focused mainly on the proximal 
cryptic 3’SSs used significantly more often in the SF3B1 
mutants (last column).

```python
def dexseq_table_entries(fn):
    """
    Return list of [number Gencode jxns tested, number novel jxns tested,
    number jxns significnat, number novel jxns significant, number cryptic
    3'SS significant, number proximal cryptic 3'SS significant, number 
    proximal cryptic 3'SS significant and use more highly in mutants]
    
    """
    df = pd.read_table(fn, index_col=0, header=0)
    out = []
    tested = df.ix[df.testable]
    gencode_tested, novel_tested = tested.ext_annotated.value_counts()
    out.append(gencode_tested)
    out.append(novel_tested)
    sig = tested[tested.padjust < ds.dexseq_p_cutoff]
    out.append(sig.shape[0])
    out.append(sig[sig.ext_annotated == False].shape[0])    
    novel_acceptor = sig.novel_acceptor.value_counts()[True]
    out.append(novel_acceptor)
    proximal = sig[(sig.downstream_acceptor_dist >= 10) & 
                   (sig.downstream_acceptor_dist <= 30)]
    out.append(proximal.shape[0])
    mut_up = (proximal['log2fold(MUT/WT)'] < 0).value_counts()[False]
    out.append(mut_up)
    return out

file_s02 = pd.DataFrame([ dexseq_table_entries(x) for x in [ds.brca_cll_um_dexseq_results, 
                                                            ds.brca_dexseq_results,
                                                            ds.cll_dexseq_results,
                                                            ds.um_dexseq_results] ],
                         index=['BRCA, CLL, UM', 'BRCA', 'CLL', 'UM'],
                         columns=['Gencode tested', 'Novel tested', 
                                  'Significant p < {}'.format(ds.dexseq_p_cutoff),
                                  'Significant novel', 'Sig. cryptic 3\'SS',
                                  'Sig. proximal cryptic 3\'SS', 
                                  'Proximal cryptic 3\'SS, mut up'])
file_s02.apply(lambda x: ['{:,}'.format(y) for y in x ]).to_csv(
    os.path.join(outdir, 'file_s02.tsv'),
    sep='\t')
file_s02
```

## File S3

619 cryptic 3’SSs located 10-30 bp upstream of canonical 
3'SSs from joint BRCA, CLL, and UM analysis. Location of 
5' splice sites and 3'SSs are one-based coordinates that 
denote the start and end of the intron. The columns COSMIC, 
TSgene, and ncg denote whether the gene is present in COSMIC, 
TSGene, or the Network of Cancer Genes respectively. 

```python
df = pd.read_table(ds.brca_cll_um_dexseq_results,
                   index_col=0, header=0)
proximal_jxns = pd.read_table(ds.proximal_jxns,
                              squeeze=True, header=None)
proximal = df.ix[proximal_jxns]

file_s03 = proximal[['chrom', 'strand', 'gene_name']]
file_s03['5\'SS'] = proximal.donor.apply(lambda x: x.split(':')[1])
file_s03['3\'SS'] = proximal.acceptor.apply(lambda x: x.split(':')[1])
file_s03['in_frame'] = proximal.downstream_acceptor_dist % 3 == 0
file_s03 = file_s03.sort(['chrom', '5\'SS', '3\'SS'])
file_s03 = file_s03[['chrom', 'strand', '5\'SS', 
                     '3\'SS', 'in_frame', 'gene_name']]
file_s03.columns = ['chrom', 'strand', '5\'SS', '3\'SS', 'in frame', 'gene']
file_s03['COSMIC'] = file_s03.gene.apply(lambda x: x in cancer_gene_census.index)
file_s03['TSGene'] = file_s03.gene.apply(lambda x: x in ts_genes.index)
file_s03['NCG'] = file_s03.gene.apply(lambda x: x in cancer_genes.index)
file_s03.to_csv(os.path.join(outdir, 'file_s03.tsv'),
                sep='\t', index=False)
print('{} rows'.format(file_s03.shape[0]))
file_s03.head()
```

## File S5

25 significant cryptic 3’SSs located 10-30 bp upstream 
of canonical 3’SSs and used more often in SF3B1 mutants 
from CLL-only DEXSeq analysis. Location of 5’ splice 
sites and 3’SSs are one-based coordinates that denote 
the start and end of the intron. The columns COSMIC, 
TSgene, and ncg denote whether the gene is present in 
COSMIC, TSGene, or the Network of Cancer Genes 
respectively.

```python
df = pd.read_table(ds.cll_dexseq_results,
                   index_col=0, header=0)
tested = df.ix[df.testable]
sig = tested[tested.padjust < ds.dexseq_p_cutoff]
proximal = sig[(sig.downstream_acceptor_dist >= 10) & 
               (sig.downstream_acceptor_dist <= 30) &
               (sig['log2fold(MUT/WT)'] > 0)]

file_s05 = proximal[['chrom', 'strand', 'gene_name']]
file_s05['5\'SS'] = proximal.donor.apply(lambda x: x.split(':')[1])
file_s05['3\'SS'] = proximal.acceptor.apply(lambda x: x.split(':')[1])
file_s05['in_frame'] = proximal.downstream_acceptor_dist % 3 == 0
file_s05 = file_s05.sort(['chrom', '5\'SS', '3\'SS'])
file_s05 = file_s05[['chrom', 'strand', '5\'SS', 
                     '3\'SS', 'in_frame', 'gene_name']]
file_s05.columns = ['chrom', 'strand', '5\'SS', '3\'SS', 'in frame', 'gene']
file_s05['COSMIC'] = file_s05.gene.apply(lambda x: x in cancer_gene_census.index)
file_s05['TSGene'] = file_s05.gene.apply(lambda x: x in ts_genes.index)
file_s05['NCG'] = file_s05.gene.apply(lambda x: x in cancer_genes.index)
file_s05.to_csv(os.path.join(outdir, 'file_s05.tsv'),
                sep='\t', index=False)
print('{} rows'.format(file_s05.shape[0]))
file_s05.head()
```

## File S6

272 genes that are differentially expressed between 
SF3B1 mutant and wild-type samples from joint analysis 
of CLL, BRCA, and UM using DESeq2.

```python
file_s06 = pd.read_csv(ds.brca_cll_um_deseq2_results,
                       index_col=0, header=0)
file_s06 = file_s06.dropna(subset=['padj'])
file_s06 = file_s06[file_s06.padj < ds.deseq2_p_cutoff]
file_s06 = file_s06[['log2FoldChange', 'pvalue', 'padj']]
file_s06.columns = ['log2 fold change (MUT/WT)', 'p-value', 
                    'Adjusted p-value']
file_s06['Gene name'] = gene_info.ix[file_s06.index, 'gene_name']
file_s06.index.name = 'Gene ID'

file_s06.to_csv(os.path.join(outdir, 'file_s06.tsv'),
                sep='\t')
print('{} rows'.format(file_s06.shape[0]))
file_s06.head()
```
