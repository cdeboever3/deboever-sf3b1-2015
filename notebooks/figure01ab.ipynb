# Figure 1a and 1b

For Figure 1a, I'm going to plot the distribution 
of distances from novel acceptors to annotated 
acceptors for junctions whose usage differs 
significantly between *SF3B1* mutants and wild-types.

For Figure 1b, I'll plot the library-normalized coverage 
as a $z$-score for significant junctions with novel 
acceptors that are 10-30 bp upstream of annotated acceptors.

```python
import brewer2mpl
import os

import ds2014 as ds
import matplotlib
import matplotlib.gridspec as gridspec
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
from matplotlib.transforms import Bbox
import pandas as pd
import scipy.spatial.distance as dist
import scipy.cluster.hierarchy as sch
```

```python
rcParams.update({'font.size': 7})
rcParams['ps.usedistiller'] = 'xpdf'
matplotlib.rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
matplotlib.rc('text', usetex=True)
```

```python
curdir = os.getcwd()
outdir = os.path.join(os.path.split(curdir)[0], 
                      'output', 
                      'figure01ab')
ds.makedir(outdir)

set1 = brewer2mpl.get_map('Set1', 'Qualitative', 9)
mut_color = set1.mpl_colors[0]
wt_color = set1.mpl_colors[1]

meta = pd.read_table(ds.metadata, index_col=0, header=0)
meta = meta[meta.disease_abbr.apply(lambda x: x in ['BRCA', 'CLL', 'UM'])]

counts = pd.read_table(ds.brca_cll_um_sj_counts,
                       index_col=0,
                       header=0)

results = pd.read_table(ds.brca_cll_um_dexseq_results,
                        index_col=0,
                        header=0)

mut_info = pd.read_table(ds.sf3b1_mut_info, header=0, index_col=0)
mut_info = mut_info[mut_info.disease_abbr.apply(lambda x: x in ['CLL', 'BRCA', 'UM'])]
mut_info.index = mut_info.name

gene_counts = pd.read_table(ds.gene_eff_counts,
                            index_col=0,
                            header=0)

size_factors = pd.read_table(ds.brca_cll_um_size_factors,
                             index_col=0,
                             header=0)
```

## Figure 1a

```python
def get_log_dists(res):
    """Get distance to novel acceptors for plotting"""
    df = res[res.novel_acceptor]
    t = log2(df.ix[((df.padjust < ds.dexseq_p_cutoff) & 
                    (df['log2fold(WT/MUT)'] < 0)),
                    'downstream_acceptor_dist'].dropna())
    t2 = log2(df.ix[((df.padjust < ds.dexseq_p_cutoff) &
                     (df['log2fold(WT/MUT)'] < 0)),
                     'upstream_acceptor_dist'].dropna())
    s = set(t.index) & set(t2.index)
    t2 = t2.drop(s)
    mut_dists = pd.concat([-1 * t, t2])
    t = log2(df.ix[((df.padjust < ds.dexseq_p_cutoff) &
                    (df['log2fold(WT/MUT)'] > 0)),
                    'downstream_acceptor_dist'].dropna())
    t2 = log2(df.ix[((df.padjust < ds.dexseq_p_cutoff) &
                     (df['log2fold(WT/MUT)'] > 0)),
                     'upstream_acceptor_dist'].dropna())
    s = set(t.index) & set(t2.index)
    t2 = t2.drop(s)
    wt_dists = pd.concat([-1 * t, t2])
    return mut_dists, wt_dists

def plot_hist(res, ax, letter_label):
    mut_dists, wt_dists = get_log_dists(res)
    if mut_dists.shape[0] > 0:
        ax.hist(mut_dists, bins=arange(-20, 21.25, 0.25),
                alpha=0.5, label='MUT$\uparrow$',
                histtype='stepfilled', linewidth=0,
                fc=mut_color)
    if wt_dists.shape[0] > 0:
        ax.hist(wt_dists, bins=arange(-20, 21.25, 0.25),
                alpha=0.5, label='WT$\uparrow$',
                histtype='stepfilled', linewidth=0,
                fc=wt_color)
    x = ax.set_xlabel('$\log_2$ distance (bp) from associated canonical acceptor')
    ax.set_ylabel('Number novel acceptors')
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    for l in ax.get_xticklines() + ax.get_yticklines(): 
        l.set_markersize(0)
    lgd = legend(loc='upper left',frameon=False)
    
    ax, ax2 = plot_polygon(ax)
    plot_little_hist(res, ax2)
    
    offset = 0.14 * (ax.get_xlim()[1] - ax.get_xlim()[0])
    t = ax.text(ax.get_xlim()[0] - offset, 
                ax.get_ylim()[1], 
                letter_label, weight='bold', 
                size=ds.subfigure_label_fontsize)
    return ax

def plot_polygon(ax):
    ymin, ymax = ax.get_ylim()
    xy = [[-1 * log2(50), 0],
          [5, ymax * 0.35],
          [20, ymax * 0.35],
          [0, 0]]
    poly = Polygon(xy, facecolor='grey',
               edgecolor='none', alpha=0.25)
    ax.add_patch(poly)
    bb_data = Bbox.from_bounds(5, ymax * 0.35, 15, ymax * 0.55)
    disp_coords = ax.transData.transform(bb_data)
    fig_coords = fig.transFigure.inverted().transform(disp_coords)
    ax2 = fig.add_axes(Bbox(fig_coords))
    return ax, ax2

def plot_little_hist(res, ax):
    df = res[res.novel_acceptor]
    t = -1 * df.ix[((df.padjust < ds.dexseq_p_cutoff) & 
                    (df['log2fold(WT/MUT)'] < 0)),
                    'downstream_acceptor_dist'].dropna()
    ax.hist(t[(t <= 0) & (t >= -50)],
            bins=range(-50, 1), alpha=0.5, 
            histtype='stepfilled', linewidth=0,
            fc=mut_color)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    for l in ax.get_xticklines() + ax.get_yticklines(): 
        l.set_markersize(0)
```

```python
fig = plt.figure(figsize=(4, 1.5))
ax = fig.add_subplot(111)
ax = plot_hist(results, ax, 'a')
fig.savefig(os.path.join(outdir, 'figure01a.pdf'),
            transparent=True, dpi=600)
```

## Figure 1b

I need to get the junctions whose usage differs 
significantly between the mutant and wild-type samples.
I will normalize the counts for these junctions by the 
size factors from DEXSeq, take $log_2$, and make a 
$z$-score.

```python
df = results[(results.padjust < ds.dexseq_p_cutoff) &
             (results.downstream_acceptor_dist <= 30) &
             (results.downstream_acceptor_dist >= 10) &
             (results['log2fold(MUT/WT)'] > 0)]

dfc = counts.ix[df.index].T
for ind in dfc.index:
    dfc.ix[ind] = dfc.ix[ind] / size_factors.ix[ind].values[0]
dfc = log2(dfc + 1)
dfc = dfc - dfc.mean()
dfc = dfc / dfc.std()


# cluster rows
dd = dist.squareform(dist.pdist(dfc))
hc = sch.linkage(dd, method='single')
dm = sch.dendrogram(hc, color_threshold=np.inf,
                    no_plot=True)
 # list mapping old indices to new indices
idx = dm['leaves']
# update data according to new indices
dfc = dfc.ix[idx] 
# cluster columns
dd = dist.squareform(dist.pdist(dfc.T))
hc = sch.linkage(dd, method='single')
dm = sch.dendrogram(hc, color_threshold=np.inf,
                    no_plot=True)
idx = dm['leaves']
dfc = dfc.ix[:,idx]

df = df.ix[dfc.columns]

# make norm
vmin = np.floor(dfc.min().min())
vmax = np.ceil(dfc.max().max())
vmax = max([vmax, abs(vmin)])
vmin = vmax * -1
my_norm = mpl.colors.Normalize(vmin, vmax)

d = dict(zip(mut_info.name, mut_info.HEAT))
meta['HEAT 5-9'] = False
#for i in meta[(meta.condition == 'MUT') &
#              (meta.disease_abbr.apply(lambda x: x in ['BRCA', 'CLL', 'UM']))].index:
for i in meta[meta.condition == 'MUT'].index:
    meta.ix[i, 'HEAT 5-9'] = d[i]
    
meta_color = ds.make_color_panel(meta)
meta_color = meta_color.ix[['condition', 'disease_abbr', 'HEAT 5-9'], 
                           dfc.index, :]

gc = pd.concat(
    [log10((gene_counts.ix[df.gene_id, meta[meta.disease_abbr == 'UM'].index] + 1).mean(axis=1)),
     log10((gene_counts.ix[df.gene_id, meta[meta.disease_abbr == 'CLL'].index] + 1).mean(axis=1)),
     log10((gene_counts.ix[df.gene_id, meta[meta.disease_abbr == 'BRCA'].index] + 1).mean(axis=1))],
    axis=1)
gc.columns = ['BRCA','CLL','UM']
gc = (gc.T / gc.max(axis=1)).T
```

```python
fig = plt.figure(figsize=(7.2, 3), dpi=600)
gs = gridspec.GridSpec(4, 4, 
                       width_ratios=[0.1, 0.05, 1, 0.025],
                       height_ratios=[0.05, 1, 0.05, 0.15])

### heatmap axis
heatmapAX = fig.add_subplot(gs[1,2])
im = heatmapAX.imshow(dfc, aspect='auto', 
                      origin='lower', 
                      interpolation='nearest',
                      norm=my_norm, cmap=cm.seismic)
for l in heatmapAX.get_xticklines() + heatmapAX.get_yticklines(): 
    l.set_markersize(0)
for sp in heatmapAX.spines.values():
    sp.set_visible(False)
ds.clean_axis(heatmapAX)

### mutant allele frequency axis
freqAX = fig.add_subplot(gs[1,0])
bL = freqAX.barh(np.arange(dfc.shape[0]) + 0.5, 
                 mut_info.alt_freq[dfc.index].replace(np.nan, 0), 
                 height=0.8, color='grey', align='center')
for t in bL:
    t.set_linewidth(0)
freqAX.set_xlim(1.01, -0.01)
freqAX.set_ylim(0, dfc.shape[0])
ds.clean_axis(freqAX)
freqAX.set_xticks([0, 0.5, 1])
freqAX.set_xticklabels([])
for l in freqAX.get_xticklines(): 
    l.set_markersize(0)
freqAX.grid(axis='x',alpha=0.5)

### freq label axis
freq_labelAX = fig.add_subplot(gs[0,0])
ds.clean_axis(freq_labelAX)
freq_labelAX.text(1, 0, '0', ha='center')
freq_labelAX.text(0.5, 0, '0.5', ha='center')
freq_labelAX.text(0,0, '1', ha='center')
freq_labelAX.text(0.5, 1.1, 'Frequency', ha='center')

### colorbar axis
colorbarAX = fig.add_subplot(gs[1,3])
cb = colorbar(mappable=im, cax=colorbarAX)
cb.solids.set_edgecolor("face")
cb.outline.set_linewidth(0)
cb.set_ticks([ int(x) for x in arange(my_norm.vmin, my_norm.vmax + 1) ])
for l in colorbarAX.get_yticklines():
    l.set_markersize(0)
cb.set_label('$log_{2}$ normalized counts $z$-score',
             rotation=270, labelpad=10)

### sample colorbar axis
sample_colorbarAX = fig.add_subplot(gs[1,1])
im = sample_colorbarAX.imshow(meta_color.transpose(1,0,2),
                              interpolation='nearest', origin='lower',
                              aspect='auto')
ds.clean_axis(sample_colorbarAX)

### plot out of frame heatmap
frameAX = fig.add_subplot(gs[2,2])
im = frameAX.imshow([df.downstream_acceptor_dist.apply(lambda x: [0,1][x % 3 != 0])],
                    aspect='auto', origin='lower', cmap=cm.Greys, interpolation='nearest')
                    
ds.clean_axis(frameAX)

### out of frame label
# make y-axis labels for out of frame heatmap. 
# Due to tight layout, it's easier for me to
# make these as custom text in the adjacent axis
frame_labelAX = fig.add_subplot(gs[2, 1])
ds.clean_axis(frame_labelAX)
t = plt.text(1, 0.5, 'Out of frame', 
             ha='right', va='center')

### plot gene expression heatmap
expAX = fig.add_subplot(gs[3,2])
im = expAX.imshow(gc.T, aspect='auto', 
                  origin='lower', interpolation='nearest',
                  cmap=cm.Greens)
ds.clean_axis(expAX)

### gene expression heatmap labels
# due to tight layout, it's easier for me to make these as custom text in the adjacent axis
exp_labelAX = fig.add_subplot(gs[3, 1])
ds.clean_axis(exp_labelAX)
t = plt.text(1, 0.85, 'BRCA',
             ha='right', va='center')
t = plt.text(1, 0.5, 'CLL',
             ha='right', va='center')
t = plt.text(1, 0.15,'UM', 
             ha='right', va='center')

### gene expression colorbar axis
exp_colorbarAX = fig.add_subplot(gs[3, 3])
cb = colorbar(mappable=im, cax=exp_colorbarAX)
cb.solids.set_edgecolor("face")
cb.outline.set_linewidth(0)
cb.set_ticks([0, 0.5, 1])
cb.set_ticklabels(['0', '0.5', '1'])
for l in exp_colorbarAX.get_yticklines():
    l.set_markersize(0)
cb.set_label('Relative\nexpression',
             multialignment='center',
             rotation=270, labelpad=14)

### legend axis
legendAX = plt.subplot(gs[0,1:])
ds.clean_axis(legendAX)
rectL = []
labelL = []
for c in ['condition', 'disease_abbr']:
    d = dict(zip(meta.ix[meta_color.major_axis,c], 
                 meta_color.ix[c].values))
    for key in d.keys():
        if key != 'NA':
            labelL.append(key.replace('MUT','\\textit{SF3B1} mutant').replace('WT','\\textit{SF3B1} wild-type'))
            #labelL.append(key.replace('MUT', 'SF3B1 mutant').replace('WT', 'SF3B1 wild-type'))
            r = Rectangle((0, 0), 0, 0, fc=d[key])
            rectL.append(r)
labelL.append('HEAT 5-9')
r = Rectangle((0, 0), 0, 0, 
              fc=meta_color.ix['HEAT 5-9', meta[meta['HEAT 5-9']].index[0]].values)
rectL.append(r)
lgd = legendAX.legend(rectL,labelL,loc='center',frameon=False,prop={'size':7},ncol=8)
for p in lgd.get_patches():
    p.set_linewidth(0)

gs.tight_layout(fig,w_pad=0.5)
fig.savefig(os.path.join(outdir, 'figure01b.pdf'),
            transparent=True, dpi=600)
```
