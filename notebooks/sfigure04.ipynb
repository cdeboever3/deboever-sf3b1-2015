# Figure S4

I'll plot the library-normalized coverage 
as a $z$-score for significant junctions with cryptic 
3'SSs that are 10-30 bp upstream of canonical 3'SSs
for the joint BRCA, CLL, LUAD, LUSC, and UM analysis.

```python
import brewer2mpl
import os

import ds2014 as ds
import matplotlib
import matplotlib.gridspec as gridspec
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
from matplotlib.transforms import Bbox
import pandas as pd
import scipy.spatial.distance as dist
import scipy.cluster.hierarchy as sch
```

```python
rcParams.update({'font.size': 7})
rcParams['ps.usedistiller'] = 'xpdf'
matplotlib.rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
matplotlib.rc('text', usetex=True)
```

```python
curdir = os.getcwd()
outdir = os.path.join(os.path.split(curdir)[0], 
                      'output', 
                      'sfigure04')
ds.makedir(outdir)

set1 = brewer2mpl.get_map('Set1', 'Qualitative', 9)
mut_color = set1.mpl_colors[0]
wt_color = set1.mpl_colors[1]

meta = pd.read_table(ds.metadata, index_col=0, header=0)

counts = pd.read_table(ds.brca_cll_luad_lusc_um_sj_counts,
                       index_col=0,
                       header=0)

results = pd.read_table(ds.brca_cll_luad_lusc_um_dexseq_results,
                        index_col=0,
                        header=0)

mut_info = pd.read_table(ds.sf3b1_mut_info, header=0, index_col=0)

size_factors = pd.read_table(ds.brca_cll_luad_lusc_um_dexseq_size_factors,
                             index_col=0,
                             header=0)
```

We are going to use the LUAD and LUSC samples here.
Only one of these samples actually has a HEAT 5-9 mutation.
One sample has two mutations in *SF3B1* but neither is
in the HEAT repeats. I'll remove one of the rows for this 
sample from the mutation DataFrame to make things easier below.

```python
mut_info = mut_info.drop_duplicates('name')
mut_info.index = mut_info.name
```

I need to get the junctions whose usage differs 
significantly between the mutant and wild-type samples.
I will normalize the counts for these junctions by the 
size factors from DEXSeq, take $log_2$, and make a 
$z$-score.

```python
df = results[(results.padjust < ds.dexseq_p_cutoff) &
             (results.downstream_acceptor_dist <= 30) &
             (results.downstream_acceptor_dist >= 10) &
             (results['log2fold(MUT/WT)'] > 0)]

dfc = counts.ix[df.index].T
for ind in dfc.index:
    dfc.ix[ind] = dfc.ix[ind] / size_factors.ix[ind].values[0]
dfc = log2(dfc + 1)
dfc = dfc - dfc.mean()
dfc = dfc / dfc.std()


# cluster rows
dd = dist.squareform(dist.pdist(dfc))
hc = sch.linkage(dd, method='complete')
dm = sch.dendrogram(hc, color_threshold=np.inf,
                    no_plot=True)
 # list mapping old indices to new indices
idx = dm['leaves']
# update data according to new indices
dfc = dfc.ix[idx] 
# cluster columns
dd = dist.squareform(dist.pdist(dfc.T))
hc = sch.linkage(dd, method='complete')
dm = sch.dendrogram(hc, color_threshold=np.inf,
                    no_plot=True)
idx = dm['leaves']
dfc = dfc.ix[:,idx]

df = df.ix[dfc.columns]

# make norm
vmin = np.floor(dfc.min().min())
vmax = np.ceil(dfc.max().max())
vmax = max([vmax, abs(vmin)])
vmin = vmax * -1
my_norm = mpl.colors.Normalize(vmin, vmax)

d = dict(zip(mut_info.name, mut_info.HEAT))
meta['HEAT 5-9'] = False
for i in meta[meta.condition == 'MUT'].index:
    meta.ix[i, 'HEAT 5-9'] = d[i]
    
meta_color = ds.make_color_panel(meta)
meta_color = meta_color.ix[['condition', 'disease_abbr', 'HEAT 5-9'], 
                           dfc.index, :]
```

I used complete as the clustering method here
rather than single as for Figure 1b. The single 
clustering splits the mutants into two branches
in opposite sides of the heatmap but complete
groups them all together. I'm clustering more for
the sake of easy visualization for the reader (it's
obvious the mutants are different from the wild-type)
so I think that makes sense.

```python
fig = plt.figure(figsize=(8, 4.1), dpi=600)
gs = gridspec.GridSpec(5, 4, 
                       width_ratios=[0.1, 0.05, 1, 0.025],
                       height_ratios=[0.05, 0.05, 1, 0.05, 0.15])

ax = fig.add_subplot(gs[0, :])
ds.clean_axis(ax)
ax.text(0, 0, 'Figure S4',
        size=16, va='bottom')

### heatmap axis
heatmapAX = fig.add_subplot(gs[2,2])
im = heatmapAX.imshow(dfc, aspect='auto', 
                      origin='lower', 
                      interpolation='nearest',
                      norm=my_norm, cmap=cm.seismic)
print('{:,} columns in heatmap'.format(dfc.shape[1]))
for l in heatmapAX.get_xticklines() + heatmapAX.get_yticklines(): 
    l.set_markersize(0)
for sp in heatmapAX.spines.values():
    sp.set_visible(False)
ds.clean_axis(heatmapAX)

### mutant allele frequency axis
freqAX = fig.add_subplot(gs[2,0])
bL = freqAX.barh(np.arange(dfc.shape[0]) + 0.5, 
                 mut_info.ix[dfc.index,'alt_freq'].replace(np.nan,0), 
                 height=0.8, color='grey', align='center')
for t in bL:
    t.set_linewidth(0)
freqAX.set_xlim(1.01, -0.01)
freqAX.set_ylim(0, dfc.shape[0])
ds.clean_axis(freqAX)
freqAX.set_xticks([0, 0.5, 1])
freqAX.set_xticklabels([])
for l in freqAX.get_xticklines(): 
    l.set_markersize(0)
freqAX.grid(axis='x',alpha=0.5)

### freq label axis
freq_labelAX = fig.add_subplot(gs[1,0])
ds.clean_axis(freq_labelAX)
freq_labelAX.text(1, 0, '0', ha='center')
freq_labelAX.text(0.5, 0, '0.5', ha='center')
freq_labelAX.text(0,0, '1', ha='center')
freq_labelAX.text(0.5, 1.1, 'Frequency', ha='center')

### colorbar axis
colorbarAX = fig.add_subplot(gs[2,3])
cb = colorbar(mappable=im, cax=colorbarAX)
cb.solids.set_edgecolor("face")
cb.outline.set_linewidth(0)
cb.set_ticks([ int(x) for x in arange(my_norm.vmin, my_norm.vmax + 1) ])
for l in colorbarAX.get_yticklines():
    l.set_markersize(0)
cb.set_label('$log_{2}$ normalized counts $z$-score',
             rotation=270, labelpad=10)

### sample colorbar axis
sample_colorbarAX = fig.add_subplot(gs[2,1])
im = sample_colorbarAX.imshow(meta_color.transpose(1,0,2),
                              interpolation='nearest', origin='lower',
                              aspect='auto')
ds.clean_axis(sample_colorbarAX)

### plot out of frame heatmap
frameAX = fig.add_subplot(gs[3,2])
im = frameAX.imshow([df.downstream_acceptor_dist.apply(lambda x: [0,1][x % 3 != 0])],
                    aspect='auto', origin='lower', cmap=cm.Greys, interpolation='nearest')
                    
ds.clean_axis(frameAX)

### out of frame label
# make y-axis labels for out of frame heatmap. 
# Due to tight layout, it's easier for me to
# make these as custom text in the adjacent axis
frame_labelAX = fig.add_subplot(gs[3, 1])
ds.clean_axis(frame_labelAX)
t = plt.text(1, 0.5, 'Out of frame', 
             ha='right', va='center')

### legend axis
legendAX = plt.subplot(gs[1,1:])
ds.clean_axis(legendAX)
rectL = []
labelL = []
for c in ['condition', 'disease_abbr']:
    d = dict(zip(meta.ix[meta_color.major_axis,c], 
                 meta_color.ix[c].values))
    for key in d.keys():
        if key != 'NA':
            labelL.append(key.replace('MUT','\\textit{SF3B1} mutant').replace('WT','\\textit{SF3B1} wild-type'))
            r = Rectangle((0, 0), 0, 0, fc=d[key])
            rectL.append(r)
labelL.append('HEAT 5-9')
r = Rectangle((0, 0), 0, 0, 
              fc=meta_color.ix['HEAT 5-9', meta[meta['HEAT 5-9']].index[0]].values)
rectL.append(r)
lgd = legendAX.legend(rectL,labelL,loc='center',frameon=False,prop={'size':7},ncol=8)
for p in lgd.get_patches():
    p.set_linewidth(0)

gs.tight_layout(fig,w_pad=0.5)
fig.savefig(os.path.join(outdir, 'sfigure04.pdf'),
            transparent=True, dpi=600)
fig.savefig(os.path.join(outdir, 'sfigure04.eps'),
            transparent=True)
fig.savefig(os.path.join(outdir, 'sfigure04.png'),
            dpi=600)
```
