```python
curdir = os.getcwd()
outdir = os.path.join(os.path.split(curdir)[0], 
                      'output', 
                      'sfigure06')
ds.makedir(outdir)

set1 = brewer2mpl.get_map('Set1', 'Qualitative', 9)
mut_color = set1.mpl_colors[0]
wt_color = set1.mpl_colors[1]

meta = pd.read_table(ds.metadata, index_col=0, header=0)

cll_psi_high = pd.read_table(ds.cll_psi_high, index_col=0,
                             header=0)

brca_psi = pd.read_table(ds.brca_psi, index_col=0,
                         header=0)

results = pd.read_table(ds.brca_dexseq_results, index_col=0,
                        header=0)

bp_analysis = pd.read_table(ds.proximal_single_bp_second, 
                            index_col=0, header=0)
```

```python
meta_color = ds.make_color_panel(meta)
c = 'condition'
d = dict(zip(meta.ix[meta_color.major_axis, c], 
             meta_color.ix[c].values))

brca_psi_high = brca_psi.ix[cll_psi_high.index].dropna()
muts = meta[(meta.condition == 'MUT') & 
            (meta.disease_abbr == 'BRCA')].index
wts = meta[(meta.condition == 'WT') & 
           (meta.disease_abbr == 'BRCA')].index
```
