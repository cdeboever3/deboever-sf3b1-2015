```python
curdir = os.getcwd()
outdir = os.path.join(os.path.split(curdir)[0], 
                      'output', 
                      'susceptible_introns')
ds.makedir(outdir)

meta = pd.read_table(ds.metadata, index_col=0, header=0)
meta = meta[meta.disease_abbr.apply(lambda x: x in ['BRCA', 'CLL', 'UM'])]

results = pd.read_table(ds.brca_cll_um_dexseq_results,
                        index_col=0,
                        header=0)

counts = pd.read_table(ds.brca_cll_um_sj_counts,
                       index_col=0,
                       header=0)

size_factors = pd.read_table(ds.brca_cll_um_dexseq_size_factors,
                             index_col=0,
                             header=0)

annot = pd.read_table(ds.brca_cll_um_sj_annot, index_col=0,
                      header=0)
```

## Get intron sequences

I'm going to get the last 50 bp of the intron for each unique
splicing acceptor. I'll go ahead and reverse complement the 
minus strand sequence.

```python
res = results.drop_duplicates(subset='acceptor')
not_sig = res[res.padjust > ds.dexseq_p_cutoff]

if not os.path.exists(ds.not_sig_intron_seq):
    plus = not_sig[not_sig.strand == '+']
    minus = not_sig[not_sig.strand == '-']

    plus = '\n'.join(plus.chrom + '\t' + plus.end.astype(str) + '\t' + 
                     plus.end.astype(str) + '\t' + plus.index + '\t0\t' + 
                     plus.strand) + '\n'
    minus = '\n'.join(minus.chrom + '\t' + (minus.start - 1).astype(str) + '\t' + 
                      (minus.start - 1).astype(str) + '\t' + minus.index + '\t0\t' + 
                      minus.strand) + '\n'

    plus = pbt.BedTool(plus, from_string=True)
    minus = pbt.BedTool(minus, from_string=True)

    plus = plus.slop(l=50, r=0, g=pbt.genome_registry.hg19)
    minus = minus.slop(l=0, r=50, g=pbt.genome_registry.hg19)
    
    plus_seq = plus.sequence(fi=ds.hg19)
    minus_seq = minus.sequence(fi=ds.hg19)

    with open(plus_seq.seqfn) as f:
        p_lines = [x.strip() for x in f.readlines()]
    with open(minus_seq.seqfn) as f:
        m_lines = [x.strip() for x in f.readlines()]

    for i in range(len(m_lines))[1::2]:
        m_lines[i] = Seq(m_lines[i]).reverse_complement().tostring()
    
    for i in range(len(p_lines))[0::2]:
        p_lines[i] += ':+'
    r = re.compile('(.+):(.+)-(.+)')    
    for i in range(len(m_lines))[0::2]:
        m = r.match(m_lines[i])
        start = int(m.group(2)) + 1
        m_lines[i] = '{}:{}-{}:-'.format(m.group(1),
                                         start,
                                         m.group(2))
    
    out = ('\n'.join(p_lines) + '\n' + 
           '\n'.join(m_lines) + '\n')

    with open(ds.not_sig_intron_seq, 'w') as f:
        f.write(out)
```

## Annotate intron sequences

```python
with open(ds.not_sig_intron_seq) as f:
    lines = [x.strip() for x in f.readlines()]
seqs = pd.DataFrame(pd.Series([x.upper() for x in lines[1::2]], 
                              index=[x[1:] for x in lines[0::2]]))
seqs.columns = ['seq']
seqs['ag'] = seqs.seq.apply(lambda x: x[-2:] == 'AG')
seqs['proximal_ag'] = seqs.seq.apply(lambda x: 'AG' in x[20:-10])
```

Let's check to make sure we obtained the correct sequences. Most 
introns should end with `AG`.

```python
seqs.seq.apply(lambda x: x[-2:]).value_counts()
```

```python
if not os.path.exists(ds.not_sig_all_bp):
    bps = ds.predict_branch_points(ds.not_sig_intron_seq)
    r = re.compile('(.+):(.+)-(.+):(.+)')
    chrom = []
    start = []
    end = []
    strand = []
    acceptor = []
    for ind in bps.index:
        m = r.match(bps.ix[ind, 'seq_id'])
        chrom.append(m.group(1))
        start.append(int(m.group(2)))
        end.append(int(m.group(3)))
        strand.append(m.group(4))
        if m.group(4) == '+':
            acceptor.append('{}:{}:{}'.format(m.group(1),
                                              m.group(3),
                                              m.group(4)))
        else:
            acceptor.append('{}:{}:{}'.format(m.group(1),
                                              m.group(2),
                                              m.group(4)))
    bps['chrom'] = chrom
    bps['start'] = start
    bps['end'] = end
    bps['strand'] = strand
    bps['acceptor'] = acceptor
    bps.to_csv(ds.not_sig_all_bp, sep='\t')
else:
    bps = pd.read_table(ds.not_sig_all_bp, index_col=0,
                        header=0)
```

## Branch points and susceptible introns

Now I'll predict branch points (BPs) for the introns and see
which introns are susceptible to cryptic splicing.

```python
uniq_bp = ds.make_uniq_bp_results(bps)
```

```python
uniq_bp.shape
```

```python
uniq_bp.head()
```

For an intron to be susceptible, it must have an AG dinucleotide 10-30 bp
upstream of the canonical 3'SS that is also 13-17 bp downstream of the 
predicted BP.

I'll want to determine some distances. I'll use
`chr1:169799786-169799836:+` as an example. I've made
the predicted branch point and potential cryptic 
3'SS lower case.
    
                bp_proximal_dist
                |-------------|
                              |-proximal_cryptic_dist|
                |--------------ss_dist---------------|
    TGTTTCTgagtaattgTATAAACATagTTGTCTTCTCTTTTATATTGCAG
                       
In this example:

* `ss_dist = 38 (attgTATAAACATagTTGTCTTCTCTTTTATATTGCAG)`
* `bp_proximal_dist = 15 (attgTATAAACATag)`
* `proximal_cryptic_dist = 23 (TTGTCTTCTCTTTTATATTGCAG)`

```python
susc = uniq_bp.ix[seqs[seqs.proximal_ag].index].dropna()
print(susc.shape)

def possible_cryptic(ind, susc=susc, seqs=seqs):
    s = seqs.ix[ind, 'seq']
    sub = s[-1 * susc.ix[ind, 'ss_dist'].astype(int):]
    return 'AG' in sub[11:17]

def cryptic_distance(ind, susc=susc, seqs=seqs):
    s = seqs.ix[ind, 'seq']
    sub = s[-1 * susc.ix[ind, 'ss_dist'].astype(int):]
    return 13 + sub[11:17].find('AG')

susc['possible_cryptic'] = [possible_cryptic(x) for x in susc.index]
susc = susc[susc.possible_cryptic]
print(susc.shape)

susc['bp_proximal_dist'] = [cryptic_distance(x) for x in susc.index]
susc['proximal_cryptic_dist'] = susc.ss_dist - susc.bp_proximal_dist

susc = susc[susc.proximal_cryptic_dist >= 10]
print(susc.shape)

def cryptic_acceptor(ind, susc=susc):
    if susc.ix[ind, 'strand'] == '+':
        out = '{}:{}:+'.format(susc.ix[ind, 'chrom'],
                               int(susc.ix[ind, 'end'] - 
                                   susc.ix[ind, 'proximal_cryptic_dist']))
    else:
        out = '{}:{}:-'.format(susc.ix[ind, 'chrom'],
                               int(susc.ix[ind, 'start'] + 
                                   susc.ix[ind, 'proximal_cryptic_dist']))
    return out

susc['cryptic_acceptor'] = [cryptic_acceptor(x) for x in susc.index]
```

## Possible cryptic SS coverage

Now that we have identified some potential crpytic junctions,
we can check whether any of them passed our coverage cutoffs
and compare the coverage between *SF3B1* mutants and wild-types.

```python
ncounts = counts.copy(deep=True).T
for ind in ncounts.index:
    ncounts.ix[ind] = ncounts.ix[ind] / size_factors.ix[ind].values[0]
ncounts = ncounts.T
```
