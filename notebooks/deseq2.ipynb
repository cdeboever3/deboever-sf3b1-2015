# DESeq2

I'm going to test for differentially expressed genes 
using the effective counts from eXpress using DESeq2.

```python
import os

import numpy as np
import pandas as pd

import ds2014 as ds

%load_ext rmagic
```

```python
curdir = os.getcwd()
outdir = os.path.join(os.path.split(curdir)[0], 
                      'output', 
                      'deseq2')
ds.makedir(outdir)

counts = pd.read_table(ds.gene_eff_counts,
                       index_col=0, header=0)
meta = pd.read_table(ds.metadata, header=0,
                     index_col=0)
brca_cll_um_dexseq = pd.read_table(ds.brca_cll_um_dexseq_results,
                                   index_col=0, header=0)

gene_info = pd.read_table(ds.gene_info, header=0,
                          index_col=0)

meta = meta[meta.disease_abbr.apply(lambda x: x in ['BRCA', 'CLL', 'UM'])]
counts = counts[counts.sum(axis=1) > 100]
counts = np.round(counts[meta.index])

meta.index = [ x.replace('-', '.') for x in meta.index ]
counts.columns = [ x.replace('-', '.') for x in counts.columns ]
index = meta.index
```

We have to change any dashes in sample names to something
else because R doesn't like dashes. I'll change them back
later. For some reason the `meta` index isn't going through
so I'm passing separately.

```python
%%R -i outdir,counts,meta,index -o sf

library(DESeq2)
rownames(meta) = index

# make DESeq data set
dds = DESeqDataSetFromMatrix(countData = counts, colData = meta, design = ~ disease_abbr + condition)

# estimate size factors and write to file
dds = estimateSizeFactors(dds)
sf = as.data.frame(sizeFactors(dds))
colnames(sf) = "size_factors"
rownames(sf) = rownames(meta)
write.table(sf, paste(outdir, "brca_cll_um_size_factors.tsv", sep="/"), quote=F, sep="\t", col.names=NA)

# estimate dispersions and plot results
dds = estimateDispersions(dds, fitType="local")
plotDispEsts(dds)

# test for differential expression
dds = nbinomWaldTest(dds)
res = results(dds)
plotMA(dds)

# write results to file
write.csv(as.data.frame(res), file=paste(outdir, "brca_cll_um_results.csv", sep="/"))

# save image
save.image(paste(outdir, "deseq2.RData", sep="/"))
```

```python
meta.index = [ x.replace('.', '-') for x in meta.index ]
counts.columns = [ x.replace('.', '-') for x in counts.columns ]

size_factors = pd.read_table(ds.brca_cll_um_deseq2_size_factors,
                             index_col=0, header=0)
size_factors.index = [x.replace('.', '-') for x in size_factors.index]
size_factors.to_csv(ds.brca_cll_um_deseq2_size_factors, sep='\t')
```

Print a list of the genes with significant gene expression
differences for use with GSEA.

```python
f = ds.brca_cll_um_deseq2_sig_genes
if not os.path.exists(f):
    brca_cll_um_results = pd.read_csv(ds.brca_cll_um_deseq2_results, 
                                      index_col=0, header=0)
    sig = brca_cll_um_results[brca_cll_um_results.padj < ds.deseq2_p_cutoff]
    gene_info.ix[sig.index, 'gene_name'].drop_duplicates().to_csv(ds.brca_cll_um_deseq2_sig_genes,
                                                                  index=False)
```
