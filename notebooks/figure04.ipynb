# Figure 4

For A, I want to plot the percent spliced in for
the proximal novel junctions in just the CLL samples.
I'll focus on just the CLL samples because the coverage 
is higher and I can avoid problems when a gene isn't 
expressed in a particular cancer.

For B, I want to make boxplots of the percent spliced in
for a subset of junctions that show a stong swith to the 
novel junction in SF3B1 mutants.

```python
import brewer2mpl
import os

import ds2014 as ds
import matplotlib
from matplotlib.colors import LogNorm
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
import pandas as pd
import scipy.spatial.distance as dist
import scipy.cluster.hierarchy as sch
```

```python
rcParams.update({'font.size': 7})
rcParams['ps.usedistiller'] = 'xpdf'
matplotlib.rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
matplotlib.rc('text', usetex=True)
```

```python
curdir = os.getcwd()
outdir = os.path.join(os.path.split(curdir)[0], 
                      'output', 
                      'figure04')
ds.makedir(outdir)

set1 = brewer2mpl.get_map('Set1', 'Qualitative', 9)
mut_color = set1.mpl_colors[0]
wt_color = set1.mpl_colors[1]

meta = pd.read_table(ds.metadata, index_col=0, header=0)

counts = pd.read_table(ds.cll_sj_counts,
                       index_col=0,
                       header=0)

results = pd.read_table(ds.cll_dexseq_results,
                        index_col=0,
                        header=0)

size_factors = pd.read_table(ds.cll_size_factors,
                             index_col=0,
                             header=0)

ighv_zap70 = pd.read_table(ds.ighv_zap70,
                           index_col=0,
                           header=0)
```

```python
def get_annotated_jxn(row):
    """
    Define the downstream annotated junction for a 
    junction with a novel acceptor using a row
    from the results table
    """
    if row['strand'] == '+':        
        return '{}:{}-{}:{}'.format(row['chrom'], 
                                    row['start'],
                                    row['end'] + int(row['downstream_acceptor_dist']),
                                    row['strand'])
    if row['strand'] == '-':
        return '{}:{}-{}:{}'.format(row['chrom'], 
                                    row['start'] - int(row['downstream_acceptor_dist']),
                                    row['end'],
                                    row['strand'])
```

```python
# A

df = results[(results.padjust < ds.dexseq_p_cutoff) &
             (results.downstream_acceptor_dist <= 30) &
             (results.downstream_acceptor_dist >= 10) &
             (results['log2fold(MUT/WT)'] > 0)]

annot_jxns = pd.Series(results.ix[df.index].apply(lambda x: get_annotated_jxn(x), axis=1))
psi = counts.ix[annot_jxns.index].values / (counts.ix[annot_jxns.index].values + 
                                            counts.ix[annot_jxns.values].values)
psi = pd.DataFrame(psi, index=annot_jxns.index,
                   columns=counts.columns)
psi = psi.dropna(how='all')
psi = psi.replace(np.nan, 0)
psi_log = log10((psi + psi.replace(0, psi.max().max() + 1).min().min()).T)
psi = psi.T

# cluster rows
dd = dist.squareform(dist.pdist(psi_log))
hc = sch.linkage(dd, method='single')
dm = sch.dendrogram(hc, color_threshold=np.inf,
                    no_plot=True)

# list mapping old indices to new indices
idx = dm['leaves']
# update data according to new indices
psi = psi.ix[idx]
psi_log = psi_log.ix[idx]

# cluster columns
dd = dist.squareform(dist.pdist(psi_log.T))
hc = sch.linkage(dd, method='single')
dm = sch.dendrogram(hc, color_threshold=np.inf,
                    no_plot=True)
idx = dm['leaves']
psi = psi.ix[:, idx]
psi_log = psi_log.ix[:, idx]

meta = ighv_zap70.join(meta.condition)
meta_color = ds.make_color_panel(meta)
meta_color = meta_color.ix[['condition', 'IGHV', 'ZAP70'], 
                           psi.index, :]

# B

cll_muts = meta[meta.condition == 'MUT'].index
cll_wts = meta[meta.condition == 'WT'].index
psi_high = psi.T[psi.ix[cll_muts].median() > 0.5]
psi_high = psi_high[psi_high[cll_wts].median(axis=1) < 0.2]
psi_high = psi_high[counts.ix[psi_high.index, cll_muts].mean(axis=1) > 30]
psi_high = psi_high[results.ix[psi_high.index, 'downstream_acceptor_dist'] % 3 != 0]
```

```python
fig = plt.figure(figsize=(5.5, 3.25),dpi=600)
gs = gridspec.GridSpec(3, 3, 
                       width_ratios=[0.05, 1, 0.05],
                       height_ratios=[0.015, 0.37, 0.015])

### heatmap axis
heatmapAX = fig.add_subplot(gs[1, 1])
im = heatmapAX.imshow(psi, aspect='auto', 
                      origin='lower', 
                      interpolation='nearest',
                      cmap=cm.Blues,
                      norm=LogNorm())
for l in heatmapAX.get_xticklines() + heatmapAX.get_yticklines(): 
    l.set_markersize(0)
for sp in heatmapAX.spines.values():
    sp.set_visible(False)
ds.clean_axis(heatmapAX)
yticks = []
heatmapAX.set_yticks(range(psi.shape[0]))
heatmapAX.set_yticklabels(yticks)

### colorbar axis
colorbarAX = fig.add_subplot(gs[1, 2])
cb = colorbar(mappable=im, cax=colorbarAX)
cb.solids.set_edgecolor("face")
cb.outline.set_linewidth(0)
for l in colorbarAX.get_yticklines():
    l.set_markersize(0)
cb.set_label('Percent spliced in',
             rotation=270, labelpad=10)

### sample colorbar axis
sample_colorbarAX = fig.add_subplot(gs[1, 0])
im = sample_colorbarAX.imshow(meta_color.transpose(1,0,2),
                              interpolation='nearest', origin='lower',
                              aspect='auto')
ds.clean_axis(sample_colorbarAX)

### plot out of frame heatmap
frameAX = fig.add_subplot(gs[2, 1])
im = frameAX.imshow([df.downstream_acceptor_dist.apply(lambda x: [0,1][x % 3 != 0])],
                    aspect='auto', origin='lower', cmap=cm.Greys, interpolation='nearest')
                    
ds.clean_axis(frameAX)

### out of frame label
# make y-axis labels for out of frame heatmap. 
# Due to tight layout, it's easier for me to
# make these as custom text in the adjacent axis
frame_labelAX = fig.add_subplot(gs[2, 2])
ds.clean_axis(frame_labelAX)
t = plt.text(0, 0.5, 'Out of frame', 
             ha='left', va='center')

### legend axis
legendAX = plt.subplot(gs[0, :])
ds.clean_axis(legendAX)
rectL = []
labelL = []
c = 'condition'
d = dict(zip(meta.ix[meta_color.major_axis, c], 
             meta_color.ix[c].values))
for key in d.keys():
    if key != 'NA':
        labelL.append(key.replace('MUT','\\textit{SF3B1} mutant').replace('WT','\\textit{SF3B1} wild-type'))
    r = Rectangle((0, 0), 0, 0, fc=d[key])
    rectL.append(r)

c = 'IGHV'
d = dict(zip(meta.ix[meta_color.major_axis, c], 
             meta_color.ix[c].values))
for key in d.keys():
    if key != 'NA':
        t = {'unmutated':'\\textit{IGHV} unmutated',
             'mutated':'\\textit{IGHV} mutated'}[key]
        labelL.append(t)
    r = Rectangle((0, 0), 0, 0, fc=d[key])
    rectL.append(r)

c = 'ZAP70'
d = dict(zip(meta.ix[meta_color.major_axis, c], 
             meta_color.ix[c].values))
for key in d.keys():
    if key != 'NA':
        labelL.append(key.replace('positive','ZAP70 positive').replace('negative','ZAP70 negative'))
    r = Rectangle((0, 0), 0, 0, fc=d[key])
    rectL.append(r)

lgd = legendAX.legend(rectL, labelL, loc='center', frameon=False, prop={'size':5}, ncol=6)
for p in lgd.get_patches():
    p.set_linewidth(0)
gs.tight_layout(fig, rect = [0, 0.5, 1, 1])
    
# B
b_gs = gridspec.GridSpec(1, psi_high.shape[0])

for i, ind in enumerate(psi_high.index):
    ax = fig.add_subplot(b_gs[0, i])
    b = ax.boxplot([psi_high.ix[ind, cll_muts],
                    psi_high.ix[ind, cll_wts]],
                   widths=0.5)
    for part in ['boxes', 'caps', 'fliers', 'medians', 'whiskers']:
        for thing in b[part]:
            thing.set_color('black')
    ax.set_xticklabels(['M', 'W'])
    g = results.ix[ind, 'gene_name']
    if g[0:2] == 'RP' and '-' in g:
        top, bottom = g.split('-')
        top += '-'
        top = '\\textit{' + top + '}'
        top += '\n'
        bottom = '\\textit{' + bottom + '}'
        g = top + bottom
    else:
        g = '\\textit{' + g + '}'
    ax.set_title(g,
                 fontsize=7,
                 style='italic',
                 multialignment='center')
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    for l in ax.get_xticklines() + ax.get_yticklines(): 
        l.set_markersize(0)
    if i % 2 == 1:
        ax.set_axis_bgcolor('#D8D8D8')
        
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    if i == 0:
        ax.set_ylabel('Percent spliced in')
        ax.spines['left'].set_visible(True)
    else:
        ax.set_yticks([])
        ax.spines['left'].set_visible(False)
b_gs.tight_layout(fig, rect = [0, 0, 1, 0.5])

t = fig.text(0.005, 0.965, 'a', weight='bold', 
             size=ds.subfigure_label_fontsize)
t = fig.text(0.005, 0.45, 'b', weight='bold', 
             size=ds.subfigure_label_fontsize)

fig.savefig(os.path.join(outdir, 'figure04.pdf'), dpi=600)
```
