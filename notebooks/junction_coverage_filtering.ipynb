# Junction Coverage Filtering

I will explore the effect of filtering out junctions
based on their coverage across the samples on the DEXSeq
results. I want to make sure that I'm not seeing 
significant results or patterns due to our choice 
of keeping junctions that have 20 junction-spanning 
reads summed over all samples.

```python
import glob
import os

import cdpybio as cpb
import ds2014 as ds
import pandas as pd
```

```python
curdir = os.getcwd()
outdir = os.path.join(os.path.split(curdir)[0], 
                      'output', 
                      'junction_coverage_filtering')
ds.makedir(outdir)

meta = pd.read_table(ds.metadata,
                     index_col=0, header=0)
meta = meta[meta.disease_abbr.apply(lambda x: x in ['BRCA', 'CLL', 'UM'])]

counts = pd.read_table(ds.brca_cll_um_sj_counts,
                       index_col=0, header=0)
results = pd.read_table(ds.brca_cll_um_dexseq_results,
                        index_col=0, header=0)
sig = results[results.padjust < ds.dexseq_p_cutoff]
novel_3ss = sig[sig.novel_acceptor]
```

I want to choose some different coverage cutoffs and show that
the DEXSeq results don't change much. Rather than re-running
DEXSeq over and over, I'll filter junctions and see whether 
significant junctions from DEXSeq are disproportionately 
filtered out. I'll try higher sums over all samples (40, 100, etc.)
and summing over just the mutants as the reviewer suggested.

## Effect of different coverage cutoffs

```python
def evaluate_results(junctions, title, counts=counts, 
                     sig=sig, novel_3ss=novel_3ss):
    """
    junctions : junctions that we are keeping (i.e. not filtering out)
    """
    print(title)
    s = set(junctions) & set(sig.index)
    n = len(s)
    p = n / float(sig.shape[0])
    print('{:,} of {:,} ({:.2%}) significant remain'.format(n, sig.shape[0], p))
    s = set(junctions) & set(novel_3ss.index)
    n = len(s)
    p = n / float(sig.shape[0])
    print('{:,} of {:,} ({:.2%}) significant novel 3\'SS remain'.format(n, novel_3ss.shape[0], p))
    n = len(junctions)
    p = n / float(counts.shape[0])
    print('{:,} of {:,} ({:.2%}) junctions remain'.format(n, counts.shape[0], p))
    
    t = sig.ix[set(junctions) & set(sig.index)]
    t = t[t.novel_acceptor]
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax = log2(t.downstream_acceptor_dist.dropna() + 1).hist(bins=100, ax=ax)
    ax.set_title(title)
    ax.set_ylabel('Number of splice sites')
    ax.set_xlabel('$log_2$ distance from canonical 3\'SS to cryptic 3\'SS');
```

I'll first look at the effect of raising the cutoff from 20 
to 50, 75, and 100 but still sum over all samples. 

```python
s = counts.sum(axis=1)
titles = ['Summed over all samples, 50',
          'Summed over all samples, 75',
          'Summed over all samples, 100']
for i, c in enumerate([50, 75, 100]):
    junctions = list(s[s > c].index)
    evaluate_results(junctions, titles[i])
    print('')
```

We can see that the distance between cryptic and proximal 3'SSs remains
10-30 bp even when raising the coverage cutoff to 100 reads.

```python
s = counts[meta[meta.condition == 'MUT'].index].sum(axis=1)
titles = ['Summed over MUT samples, 50',
          'Summed over MUT samples, 75',
          'Summed over MUT samples, 100']
for i, c in enumerate([25, 50, 75]):
    junctions = list(s[s > c].index)
    evaluate_results(junctions, titles[i])
    print('')
```

We can see that the distance between cryptic and proximal 3'SSs remains
10-30 bp even when raising the coverage cutoff to 100 reads summed over
only the mutants.
