# Supplementary Tables

```python
import os

import pandas as pd

import ds2014 as ds
```

```python
curdir = os.getcwd()
outdir = os.path.join(os.path.split(curdir)[0], 
                      'output', 
                      'supplementary_tables')
ds.makedir(outdir)
```

```python
meta = pd.read_table(ds.metadata, header=0,
                     index_col=0)

gene_info = pd.read_table(ds.gene_info, header=0,
                          index_col=0)

# cancer census genes from COSMIC
cancer_gene_census = pd.read_table(ds.cancer_gene_census,
                                   index_col=0)
# cancer genes from NCG4.0 (http://ncg.kcl.ac.uk/)
cancer_genes = pd.read_table(ds.cancer_genes,
                             index_col=1, header=0)
# tumor suppressors from TSGene (http://bioinfo.mc.vanderbilt.edu/TSGene/)
ts_genes = pd.read_table(ds.ts_genes,
                         index_col=1, header=0)
```

## Supplementary Table 1

```python
sf3b1_mut_info = pd.read_table(ds.sf3b1_mut_info,
                               header=0, index_col=None)
stable01 = sf3b1_mut_info[['name', 'disease_abbr',
                           'alt_freq', 'mut_type',
                           'codon_change', 'HEAT']]
stable01['alt_freq'] = stable01.alt_freq.apply(lambda x: '{:.2f}'.format(x)).values
stable01.columns = ['Sample Name', 'Cancer', 'Alternate Allele Frequency',
                    'Mutation Type', 'Codon Change', 'In HEAT 5-9']
stable01['SF3B1 mutated'] = True
wt = pd.DataFrame(columns=stable01.columns)
wt['Sample Name'] = meta[meta.condition == 'WT'].index
wt['SF3B1 mutated'] = False
wt['Cancer'] = meta.ix[wt['Sample Name'], 'disease_abbr'].values
stable01 = pd.concat([stable01, wt])
stable01.to_csv(os.path.join(outdir, 'stable01.tsv'),
                sep='\t', index=False)
stable01.head()
```

## Supplementary Table 2

```python
def dexseq_table_entries(fn):
    """
    Return list of [number Gencode jxns tested, number novel jxns tested,
    number jxns significnat, number novel jxns significant, number cryptic
    3'SS significant, number proximal cryptic 3'SS significant, number 
    proximal cryptic 3'SS significant and use more highly in mutants]
    
    """
    df = pd.read_table(fn, index_col=0, header=0)
    out = []
    tested = df.ix[df.testable]
    gencode_tested, novel_tested = tested.ext_annotated.value_counts()
    out.append(gencode_tested)
    out.append(novel_tested)
    sig = tested[tested.padjust < ds.dexseq_p_cutoff]
    out.append(sig.shape[0])
    out.append(sig[sig.ext_annotated == False].shape[0])    
    novel_acceptor = sig.novel_acceptor.value_counts()[True]
    out.append(novel_acceptor)
    proximal = sig[(sig.downstream_acceptor_dist >= 10) & 
                   (sig.downstream_acceptor_dist <= 30)]
    out.append(proximal.shape[0])
    mut_up = (proximal['log2fold(MUT/WT)'] < 0).value_counts()[False]
    out.append(mut_up)
    return out

stable02 = pd.DataFrame([ dexseq_table_entries(x) for x in [ds.brca_cll_um_dexseq_results, 
                                                            ds.brca_dexseq_results,
                                                            ds.cll_dexseq_results,
                                                            ds.um_dexseq_results] ],
                         index=['BRCA, CLL, UM', 'BRCA', 'CLL', 'UM'],
                         columns=['Gencode tested', 'Novel tested', 
                                  'Significant p < {}'.format(ds.dexseq_p_cutoff),
                                  'Significant novel', 'Sig. cryptic 3\'SS',
                                  'Sig. proximal cryptic 3\'SS', 
                                  'Proximal cryptic 3\'SS, mut up'])
stable02.apply(lambda x: ['{:,}'.format(y) for y in x ]).to_csv(
    os.path.join(outdir, 'stable02.tsv'),
    sep='\t')
stable02
```

## Supplementary Table 3

```python
df = pd.read_table(ds.brca_cll_um_dexseq_results,
                   index_col=0, header=0)
proximal_jxns = pd.read_table(ds.proximal_jxns,
                              squeeze=True, header=None)
proximal = df.ix[proximal_jxns]

stable03 = proximal[['chrom', 'strand', 'gene_name']]
stable03['5\'SS'] = proximal.donor.apply(lambda x: x.split(':')[1])
stable03['3\'SS'] = proximal.acceptor.apply(lambda x: x.split(':')[1])
stable03['in_frame'] = proximal.downstream_acceptor_dist % 3 == 0
stable03 = stable03.sort(['chrom', '5\'SS', '3\'SS'])
stable03 = stable03[['chrom', 'strand', '5\'SS', 
                     '3\'SS', 'in_frame', 'gene_name']]
stable03.columns = ['chrom', 'strand', '5\'SS', '3\'SS', 'in frame', 'gene']
stable03['COSMIC'] = stable03.gene.apply(lambda x: x in cancer_gene_census.index)
stable03['TSGene'] = stable03.gene.apply(lambda x: x in ts_genes.index)
stable03['NCG'] = stable03.gene.apply(lambda x: x in cancer_genes.index)
stable03.to_csv(os.path.join(outdir, 'stable03.tsv'),
                sep='\t', index=False)
stable03.head()
```

## Supplementary Table 5

```python
df = pd.read_table(ds.cll_dexseq_results,
                   index_col=0, header=0)
tested = df.ix[df.testable]
sig = tested[tested.padjust < ds.dexseq_p_cutoff]
proximal = sig[(sig.downstream_acceptor_dist >= 10) & 
               (sig.downstream_acceptor_dist <= 30) &
               (sig['log2fold(MUT/WT)'] > 0)]

stable05 = proximal[['chrom', 'strand', 'gene_name']]
stable05['5\'SS'] = proximal.donor.apply(lambda x: x.split(':')[1])
stable05['3\'SS'] = proximal.acceptor.apply(lambda x: x.split(':')[1])
stable05['in_frame'] = proximal.downstream_acceptor_dist % 3 == 0
stable05 = stable05.sort(['chrom', '5\'SS', '3\'SS'])
stable05 = stable05[['chrom', 'strand', '5\'SS', 
                     '3\'SS', 'in_frame', 'gene_name']]
stable05.columns = ['chrom', 'strand', '5\'SS', '3\'SS', 'in frame', 'gene']
stable05['COSMIC'] = stable05.gene.apply(lambda x: x in cancer_gene_census.index)
stable05['TSGene'] = stable05.gene.apply(lambda x: x in ts_genes.index)
stable05['NCG'] = stable05.gene.apply(lambda x: x in cancer_genes.index)
stable05.to_csv(os.path.join(outdir, 'stable05.tsv'),
                sep='\t', index=False)
stable05.head()
```

## Supplementary Table 6

Genes whose expression differs significantly between
*SF3B1* mutant and wild-type samples for analysis
of BRCA, CLL, and UM samples using DESeq2.

```python
stable06 = pd.read_csv(ds.brca_cll_um_deseq2_results,
                       index_col=0, header=0)
stable06 = stable06.dropna(subset=['padj'])
stable06 = stable06[stable06.padj < ds.deseq2_p_cutoff]
stable06 = stable06[['log2FoldChange', 'pvalue', 'padj']]
stable06.columns = ['log2 fold change (MUT/WT)', 'p-value', 
                    'Adjusted p-value']
stable06['Gene name'] = gene_info.ix[stable06.index, 'gene_name']
stable06.index.name = 'Gene ID'

stable06.to_csv(os.path.join(outdir, 'stable06.tsv'),
                sep='\t')

stable06.head()
```
