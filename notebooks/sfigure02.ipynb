# Supplementary Figure 2

I'm going to plot histograms of the distance from 
novel donors and acecptors that are significantly 
differently used between *SF3B1* mutants and wild-types
for each of the BRCA, CLL, and UM individual DEXSeq
analyses.

```python
from copy import deepcopy
import os

import brewer2mpl
import ds2014 as ds
import matplotlib.gridspec as gridspec
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
from matplotlib.transforms import Bbox
import pandas as pd
```

```python
curdir = os.getcwd()
outdir = os.path.join(os.path.split(curdir)[0], 
                      'output', 
                      'sfigure02')
ds.makedir(outdir)

rcParams.update({'font.size': 7})
rcParams['ps.usedistiller'] = 'xpdf'
matplotlib.rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
matplotlib.rc('text', usetex=True)

set1 = brewer2mpl.get_map('Set1', 'Qualitative', 9)
mut_color = set1.mpl_colors[0]
wt_color = set1.mpl_colors[1]

brca_results = pd.read_table(ds.brca_dexseq_results,
                             index_col=0, header=0)
cll_results = pd.read_table(ds.cll_dexseq_results,
                            index_col=0, header=0)
um_results = pd.read_table(ds.um_dexseq_results,
                           index_col=0, header=0)
```

```python
def get_log_dists(res):
    """Get distance to novel acceptors for plotting"""
    df = res[res.novel_acceptor]
    t = log2(df.ix[((df.padjust < ds.dexseq_p_cutoff) & 
                    (df['log2fold(WT/MUT)'] < 0)),
                    'downstream_acceptor_dist'].dropna())
    t2 = log2(df.ix[((df.padjust < ds.dexseq_p_cutoff) &
                     (df['log2fold(WT/MUT)'] < 0)),
                     'upstream_acceptor_dist'].dropna())
    s = set(t.index) & set(t2.index)
    t2 = t2.drop(s)
    mut_dists = pd.concat([-1 * t, t2])
    t = log2(df.ix[((df.padjust < ds.dexseq_p_cutoff) &
                    (df['log2fold(WT/MUT)'] > 0)),
                    'downstream_acceptor_dist'].dropna())
    t2 = log2(df.ix[((df.padjust < ds.dexseq_p_cutoff) &
                     (df['log2fold(WT/MUT)'] > 0)),
                     'upstream_acceptor_dist'].dropna())
    s = set(t.index) & set(t2.index)
    t2 = t2.drop(s)
    wt_dists = pd.concat([-1 * t, t2])
    return mut_dists, wt_dists

def plot_hist(res, ax, letter_label):
    mut_dists, wt_dists = get_log_dists(res)
    if mut_dists.shape[0] > 0:
        ax.hist(mut_dists, bins=arange(-20, 21.25, 0.25),
                alpha=0.5, label='MUT$\uparrow$',
                histtype='stepfilled', linewidth=0,
                fc=mut_color)
    if wt_dists.shape[0] > 0:
        ax.hist(wt_dists, bins=arange(-20, 21.25, 0.25),
                alpha=0.5, label='WT$\uparrow$',
                histtype='stepfilled', linewidth=0,
                fc=wt_color)
    x = ax.set_xlabel('$\log_2$ distance (bp) from associated canonical acceptor')
    ax.set_ylabel('Number novel acceptors')
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    for l in ax.get_xticklines() + ax.get_yticklines(): 
        l.set_markersize(0)
    lgd = ax.legend(loc='upper left', frameon=False)
    
    offset = 0.11 * (ax.get_xlim()[1] - ax.get_xlim()[0])
    t = ax.text(ax.get_xlim()[0] - offset, 
                ax.get_ylim()[1], 
                letter_label, weight='bold', 
                size=ds.subfigure_label_fontsize)
    return ax

def plot_polygon(ax):
    ymin, ymax = ax.get_ylim()
    xy = [[-1 * log2(50), 0],
          [5, ymax * 0.35],
          [20, ymax * 0.35],
          [0, 0]]
    poly = Polygon(xy, facecolor='grey',
               edgecolor='none', alpha=0.25)
    ax.add_patch(poly)
    bb_data = Bbox.from_bounds(5, ymax * 0.35, 15, ymax * 0.55)
    disp_coords = ax.transData.transform(bb_data)
    fig_coords = fig.transFigure.inverted().transform(disp_coords)
    ax2 = fig.add_axes(Bbox(fig_coords))
    return ax, ax2

def plot_little_hist(res, ax):
    df = res[res.novel_acceptor]
    t = -1 * df.ix[((df.padjust < ds.dexseq_p_cutoff) & 
                    (df['log2fold(WT/MUT)'] < 0)),
                    'downstream_acceptor_dist'].dropna()
    ax.hist(t[(t <= 0) & (t >= -50)],
            bins=range(-50, 1), alpha=0.5, 
            histtype='stepfilled', linewidth=0,
            fc=mut_color)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    for l in ax.get_xticklines() + ax.get_yticklines(): 
        l.set_markersize(0)
```

```python
fig = plt.figure(figsize=(4, 6.25))

gs = gridspec.GridSpec(4, 1,
                       height_ratios=[0.1, 0.3,
                                      0.3, 0.3])

ax = fig.add_subplot(gs[0, 0])
ds.clean_axis(ax)
ax.text(0, 0, 'Supplementary Figure 2',
        size=16, va='bottom')

axs = []
ax = fig.add_subplot(gs[1, 0])
ax = plot_hist(brca_results, ax, 'a')
axs.append(ax)

ax = fig.add_subplot(gs[2, 0])
ax = plot_hist(cll_results, ax, 'b')
axs.append(ax)

ax = fig.add_subplot(gs[3, 0])
ax = plot_hist(um_results, ax, 'c')
axs.append(ax)

fig.tight_layout()

for i, res in enumerate([brca_results, cll_results,
                         um_results]):
    ax, ax2 = plot_polygon(axs[i])
    plot_little_hist(res, ax2)

fig.savefig(os.path.join(outdir, 'sfigure02.pdf'),
            transparent=True)
```
