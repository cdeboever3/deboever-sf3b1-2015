# Figure S2

I'm going to plot histograms of the distance from 
novel donors and acecptors that are significantly 
differently used between *SF3B1* mutants and wild-types
for each of the BRCA, CLL, and UM individual DEXSeq
analyses.

```python
from copy import deepcopy
import os

import brewer2mpl
import ds2014 as ds
from ds2014.fig import get_sig_log_dists
from ds2014.fig import get_not_sig_log_dists
from ds2014.fig import plot_hist
from ds2014.fig import plot_polygon
from ds2014.fig import plot_little_hist
import matplotlib.gridspec as gridspec
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
from matplotlib.transforms import Bbox
import pandas as pd

from pylab import *
%matplotlib inline
```

```python
curdir = os.getcwd()
outdir = os.path.join(os.path.split(curdir)[0], 
                      'output', 
                      'sfigure02')
ds.makedir(outdir)

rcParams.update({'font.size': 7})
rcParams['ps.usedistiller'] = 'xpdf'
matplotlib.rc('font',**{'family':'sans-serif','sans-serif':['Arial']})
#matplotlib.rc('text', usetex=True)

set1 = brewer2mpl.get_map('Set1', 'Qualitative', 9)
mut_color = set1.mpl_colors[0]
wt_color = set1.mpl_colors[1]

brca_results = pd.read_table(ds.brca_dexseq_results,
                             index_col=0, header=0)
cll_results = pd.read_table(ds.cll_dexseq_results,
                            index_col=0, header=0)
um_results = pd.read_table(ds.um_dexseq_results,
                           index_col=0, header=0)
```

```python
fig = plt.figure(figsize=(4, 6.25))

gs = gridspec.GridSpec(4, 1,
                       height_ratios=[0.1, 0.3,
                                      0.3, 0.3])

ax = fig.add_subplot(gs[0, 0])
ds.clean_axis(ax)
ax.text(0, 0, 'Figure S2',
        size=16, va='bottom')

axs = []
ax = fig.add_subplot(gs[1, 0])
ax = plot_hist(brca_results, ax,
               mut_color, wt_color,
               letter_label='A')
axs.append(ax)

ax = fig.add_subplot(gs[2, 0])
ax = plot_hist(cll_results, ax,
               mut_color, wt_color,
               letter_label='B')
axs.append(ax)

ax = fig.add_subplot(gs[3, 0])
ax = plot_hist(um_results, ax,
               mut_color, wt_color,
               letter_label='C')
axs.append(ax)

fig.tight_layout()

for i, res in enumerate([brca_results, cll_results,
                         um_results]):
    ax, ax2 = plot_polygon(axs[i], fig)
    plot_little_hist(res, ax2, mut_color, wt_color)

fig.savefig(os.path.join(outdir, 'sfigure02.pdf'),
            transparent=True)
fig.savefig(os.path.join(outdir, 'S2 Figure.eps'),
            transparent=True)
fig.savefig(os.path.join(outdir, 'sfigure02.png'),
            dpi=600)
```
