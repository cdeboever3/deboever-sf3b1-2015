of distances from novel 3'SSs to canonical
3'SSs for junctions whose usage differs 
significantly between *SF3B1* mutants and wild-types.

as a $z$-score for significant junctions with novel

On the top I'll plot the frequency
of mutations at each amino acid from COSMIC ($log_{10}$

## Figure 1c preparation

### Format COSMIC data

I'm going to format the COSMIC data so that it's 
easier to use.

```python
def convert_aa(x):
    """Strip p. off amino acid change"""
    if type(x) == float:
        out = np.nan
    else:
        out = x.split('.')[1]
    return out

cosmic['Mutation AA'] = cosmic['Mutation AA'].apply(lambda x: convert_aa(x))

# I'm going to make a column with the codon 
# number for each amino acid. I'll leave it 
# one-based.
codonRE = re.compile('(\D*)(\d*)(\D*)')
def aa_to_codon(x):
    "Convert amino acid change to codon number"""
    if type(x) == float:
        out = np.nan
    else:
        out = int(codonRE.match(x).group(2))
    return out

cosmic['Mutation codon'] = cosmic['Mutation AA'].apply(lambda x: aa_to_codon(x))
```

### HEAT repeats definition

The dict `wang_heat` defined below shows where the first 10 HEAT repeats start according to [Wang et al. 1998](http://www.ncbi.nlm.nih.gov/pubmed/9585501). Note that these coordinates are one-based. I defined them by searching for the beginning sequence of the repeat in my protein sequence then adding one to make one-based (inclusive, exclusive). For instance, the first residue of HEAT repeat 5 is 608 and the last is 639 while the first residue of repeat 6 is 640.

It seems that there might be an error in Wang's figure because he has residue 458 labeled as Q, but that should actually be 459 in one-based coordinates. 

These HEAT repeat definitions differ from the current definitions in UniProt but the *SF3B1* mutation field seems to use Wang's definitions, so I will do that here.

```python
wang_heat = {1:459, 2:496,3:531, 4:570, 5:608, 
             6:640, 7:675, 8:724, 9:766, 10:805}
```

## Figure 1b preparation

I need to get the junctions whose usage differs 
significantly between the mutant and wild-type samples.
I will normalize the counts for these junctions by the 
size factors from DEXSeq, take $log_2$, and make a 
$z$-score.

```python
df = results[(results.padjust < ds.dexseq_p_cutoff) &
             (results.downstream_acceptor_dist <= 30) &
             (results.downstream_acceptor_dist >= 10) &
             (results['log2fold(MUT/WT)'] > 0)]

dfc = counts.ix[df.index].T
for ind in dfc.index:
    dfc.ix[ind] = dfc.ix[ind] / size_factors.ix[ind].values[0]
dfc = log2(dfc + 1)
dfc = dfc - dfc.mean()
dfc = dfc / dfc.std()


# cluster rows
dd = dist.squareform(dist.pdist(dfc))
hc = sch.linkage(dd, method='single')
dm = sch.dendrogram(hc, color_threshold=np.inf,
                    no_plot=True)
 # list mapping old indices to new indices
idx = dm['leaves']
# update data according to new indices
dfc = dfc.ix[idx] 
# cluster columns
dd = dist.squareform(dist.pdist(dfc.T))
hc = sch.linkage(dd, method='single')
dm = sch.dendrogram(hc, color_threshold=np.inf,
                    no_plot=True)
idx = dm['leaves']
dfc = dfc.ix[:,idx]

df = df.ix[dfc.columns]

# make norm
vmin = np.floor(dfc.min().min())
vmax = np.ceil(dfc.max().max())
vmax = max([vmax, abs(vmin)])
vmin = vmax * -1
my_norm = mpl.colors.Normalize(vmin, vmax)

d = dict(zip(mut_info.name, mut_info.HEAT))
meta['HEAT 5-9'] = False
for i in meta[meta.condition == 'MUT'].index:
    meta.ix[i, 'HEAT 5-9'] = d[i]
    
meta_color = ds.make_color_panel(meta)
meta_color = meta_color.ix[['condition', 'disease_abbr', 'HEAT 5-9'], 
                           dfc.index, :]

gc = pd.concat(
    [log10((gene_fpkm.ix[df.gene_id, meta[meta.disease_abbr == 'UM'].index] + 1).mean(axis=1)),
     log10((gene_fpkm.ix[df.gene_id, meta[meta.disease_abbr == 'CLL'].index] + 1).mean(axis=1)),
     log10((gene_fpkm.ix[df.gene_id, meta[meta.disease_abbr == 'BRCA'].index] + 1).mean(axis=1))],
    axis=1)
gc.columns = ['BRCA','CLL','UM']
gc = (gc.T / gc.max(axis=1)).T
```
