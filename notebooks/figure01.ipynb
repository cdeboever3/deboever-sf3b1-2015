# Figure 1

For Figure 1a, I'm going to plot the distribution 
of distances from novel acceptors to annotated 
acceptors for junctions whose usage differs 
significantly between *SF3B1* mutants and wild-types.

For Figure 1b, I'll plot the library-normalized coverage 
as a $z$-score for significant junctions with novel 
acceptors that are 10-30 bp upstream of annotated acceptors.

Figure 1c will show the frequency of various mutations in
the HEAT repeats from COSMIC and this study. I'm going to make 
the $x$-axis the HEAT repeats (e.g. the protein sequence). 
On the top I'll plot the frequency
of mutations at each amino acid from COSMIC ($log_{10}$
scale) and on the bottom I'll plot the frequency of mutations
observed in this study.

## Figure 1a

```python
def get_log_dists(res):
    """Get distance to novel acceptors for plotting"""
    df = res[res.novel_acceptor]
    t = log2(df.ix[((df.padjust < ds.dexseq_p_cutoff) & 
                    (df['log2fold(WT/MUT)'] < 0)),
                    'downstream_acceptor_dist'].dropna())
    t2 = log2(df.ix[((df.padjust < ds.dexseq_p_cutoff) &
                     (df['log2fold(WT/MUT)'] < 0)),
                     'upstream_acceptor_dist'].dropna())
    s = set(t.index) & set(t2.index)
    t2 = t2.drop(s)
    mut_dists = pd.concat([-1 * t, t2])
    t = log2(df.ix[((df.padjust < ds.dexseq_p_cutoff) &
                    (df['log2fold(WT/MUT)'] > 0)),
                    'downstream_acceptor_dist'].dropna())
    t2 = log2(df.ix[((df.padjust < ds.dexseq_p_cutoff) &
                     (df['log2fold(WT/MUT)'] > 0)),
                     'upstream_acceptor_dist'].dropna())
    s = set(t.index) & set(t2.index)
    t2 = t2.drop(s)
    wt_dists = pd.concat([-1 * t, t2])
    return mut_dists, wt_dists

def plot_hist(res, ax, letter_label=None):
    mut_dists, wt_dists = get_log_dists(res)
    if mut_dists.shape[0] > 0:
        ax.hist(mut_dists, bins=arange(-20, 21.25, 0.25),
                alpha=0.5, label='MUT$\uparrow$',
                histtype='stepfilled', linewidth=0,
                fc=mut_color)
    if wt_dists.shape[0] > 0:
        ax.hist(wt_dists, bins=arange(-20, 21.25, 0.25),
                alpha=0.5, label='WT$\uparrow$',
                histtype='stepfilled', linewidth=0,
                fc=wt_color)
    x = ax.set_xlabel('$\log_2$ distance (bp) from associated canonical acceptor')
    ax.set_ylabel('Number novel acceptors')
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    for l in ax.get_xticklines() + ax.get_yticklines(): 
        l.set_markersize(0)
    lgd = ax.legend(loc='upper left',frameon=False)
    ax.xaxis.set_major_formatter(ds.comma_format)
    ax.yaxis.set_major_formatter(ds.comma_format)
    
    if letter_label:
        offset = 0.14 * (ax.get_xlim()[1] - ax.get_xlim()[0])
        t = ax.text(ax.get_xlim()[0] - offset, 
                    ax.get_ylim()[1], 
                    letter_label, weight='bold', 
                    size=ds.subfigure_label_fontsize)
    return ax

def plot_polygon(ax):
    ymin, ymax = ax.get_ylim()
    xy = [[-1 * log2(50), 0],
          [5, ymax * 0.35],
          [20, ymax * 0.35],
          [0, 0]]
    poly = Polygon(xy, facecolor='grey',
               edgecolor='none', alpha=0.25)
    ax.add_patch(poly)
    bb_data = Bbox.from_bounds(5, ymax * 0.35, 15, ymax * 0.55)
    disp_coords = ax.transData.transform(bb_data)
    fig_coords = fig.transFigure.inverted().transform(disp_coords)
    ax2 = fig.add_axes(Bbox(fig_coords))
    return ax, ax2

def plot_little_hist(res, ax):
    df = res[res.novel_acceptor]
    t = -1 * df.ix[((df.padjust < ds.dexseq_p_cutoff) & 
                    (df['log2fold(WT/MUT)'] < 0)),
                    'downstream_acceptor_dist'].dropna()
    ax.hist(t[(t <= 0) & (t >= -50)],
            bins=range(-50, 1), alpha=0.5, 
            histtype='stepfilled', linewidth=0,
            fc=mut_color)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    for l in ax.get_xticklines() + ax.get_yticklines(): 
        l.set_markersize(0)
    ax.xaxis.set_major_formatter(ds.comma_format)
    ax.yaxis.set_major_formatter(ds.comma_format)
```

## Format COSMIC data

I'm going to format the COSMIC data so that it's 
more useful.

```python
def convert_aa(x):
    """Strip p. off amino acid change"""
    if type(x) == float:
        out = np.nan
    else:
        out = x.split('.')[1]
    return out

cosmic['Mutation AA'] = cosmic['Mutation AA'].apply(lambda x: convert_aa(x))

# I'm going to make a column with the codon 
# number for each amino acid. I'll leave it 
# one-based.
codonRE = re.compile('(\D*)(\d*)(\D*)')
def aa_to_codon(x):
    "Convert amino acid change to codon number"""
    if type(x) == float:
        out = np.nan
    else:
        out = int(codonRE.match(x).group(2))
    return out

cosmic['Mutation codon'] = cosmic['Mutation AA'].apply(lambda x: aa_to_codon(x))
```

## HEAT repeats definition

The dict `wang_heat` defined below shows where the first 10 HEAT repeats start according to [Wang et al. 1998](http://www.ncbi.nlm.nih.gov/pubmed/9585501). Note that these coordinates are one-based. I defined them by searching for the beginning sequence of the repeat in my protein sequence then adding one to make one-based (inclusive, exclusive). For instance, the first residue of HEAT repeat 5 is 608 and the last is 639 while the first residue of repeat 6 is 640.

It seems that there might be an error in Wang's figure because he has residue 458 labeled as Q, but that should actually be 459 in one-based coordinates. 

These HEAT repeat definitions differ from the current definitions in UniProt but the *SF3B1* mutation field seems to use Wang's definitions, so I will do that here.

```python
wang_heat = {1:459, 2:496,3:531, 4:570, 5:608, 6:640, 7:675, 8:724,
             9:766, 10:805}
```

## Figure 1b

I need to get the junctions whose usage differs 
significantly between the mutant and wild-type samples.
I will normalize the counts for these junctions by the 
size factors from DEXSeq, take $log_2$, and make a 
$z$-score.

```python
df = results[(results.padjust < ds.dexseq_p_cutoff) &
             (results.downstream_acceptor_dist <= 30) &
             (results.downstream_acceptor_dist >= 10) &
             (results['log2fold(MUT/WT)'] > 0)]

dfc = counts.ix[df.index].T
for ind in dfc.index:
    dfc.ix[ind] = dfc.ix[ind] / size_factors.ix[ind].values[0]
dfc = log2(dfc + 1)
dfc = dfc - dfc.mean()
dfc = dfc / dfc.std()


# cluster rows
dd = dist.squareform(dist.pdist(dfc))
hc = sch.linkage(dd, method='single')
dm = sch.dendrogram(hc, color_threshold=np.inf,
                    no_plot=True)
 # list mapping old indices to new indices
idx = dm['leaves']
# update data according to new indices
dfc = dfc.ix[idx] 
# cluster columns
dd = dist.squareform(dist.pdist(dfc.T))
hc = sch.linkage(dd, method='single')
dm = sch.dendrogram(hc, color_threshold=np.inf,
                    no_plot=True)
idx = dm['leaves']
dfc = dfc.ix[:,idx]

df = df.ix[dfc.columns]

# make norm
vmin = np.floor(dfc.min().min())
vmax = np.ceil(dfc.max().max())
vmax = max([vmax, abs(vmin)])
vmin = vmax * -1
my_norm = mpl.colors.Normalize(vmin, vmax)

d = dict(zip(mut_info.name, mut_info.HEAT))
meta['HEAT 5-9'] = False
for i in meta[meta.condition == 'MUT'].index:
    meta.ix[i, 'HEAT 5-9'] = d[i]
    
meta_color = ds.make_color_panel(meta)
meta_color = meta_color.ix[['condition', 'disease_abbr', 'HEAT 5-9'], 
                           dfc.index, :]

gc = pd.concat(
    [log10((gene_fpkm.ix[df.gene_id, meta[meta.disease_abbr == 'UM'].index] + 1).mean(axis=1)),
     log10((gene_fpkm.ix[df.gene_id, meta[meta.disease_abbr == 'CLL'].index] + 1).mean(axis=1)),
     log10((gene_fpkm.ix[df.gene_id, meta[meta.disease_abbr == 'BRCA'].index] + 1).mean(axis=1))],
    axis=1)
gc.columns = ['BRCA','CLL','UM']
gc = (gc.T / gc.max(axis=1)).T
```
